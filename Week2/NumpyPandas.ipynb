```python
df7.corr() #if correlation between x and y is more towards +1/ -1 , that means there exists direct relationship between x,y 

# for eg, y = 3x has correlation =1 
#         y = -3x has correlation =-1 
#         y = x^2 has correlation 0.97
#         y = -x^2 has correlation -0.97
# calculate corr of values online - https://www.socscistatistics.com/tests/pearson/default2.aspx 


#basically abs of corr is very important thing, helps to model the data
```

```python
corr_df7 = df7.corr() 

corr_df7.values
```

```python
# importing ploting libraries 
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
corr_df7.values
```

```python
cols = corr_df7.index
cols # cols contain name of columns in corr_df7 dataframe
```

```python
sns.set(font_scale=1.25)
hm = sns.heatmap(corr_df7, cbar=True, square=True, annot=True, annot_kws={'size': 10}, fmt='.2f', yticklabels=cols.values, xticklabels=cols.values, cmap='coolwarm')


#cbar -- > added colorbar in heatmap 
# annot --> added annotations(float values in this case) 
```

```python
# we can see that the above map is symmetrical along the diagonal; which is kind of obvious; because corr(x,y) == corr(y,x) :)
```
